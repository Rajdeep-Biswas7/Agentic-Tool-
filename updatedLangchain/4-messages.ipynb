```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("What is the capital of France?")
print(response.content)
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
messages = [
        SystemMessage(content="You are a helpful assistant."),
        HumanMessage(content="What is the capital of France?"),
        AIMessage(content="The capital of France is Paris.")
 ]
response=model.invoke(messages)
response.content
```

```python
system_msg = SystemMessage(content="You are a helpful assistant.")

messages = [
    system_msg,
    HumanMessage(content="What is the capital of France?"),
    AIMessage(content="The capital of France is Paris.")
]
response = model.invoke(messages)
print(response.content)
```

```python
## detail info to the llm through system message
system_msg = SystemMessage(content="You are a helpful assistant. Please answer the question in detail.")
messages = [
    system_msg,
    HumanMessage(content="What is the capital of France?"),
    AIMessage(content="The capital of France is Paris.")
]
response = model.invoke(messages)
print(response.content)
```

```python
##message metadata
human_msg=HumanMessage(
    content="hello",
    name="alice",
    id="12345",
)
```

```python
response= model.invoke([
    human_msg
])
response
```

```python
# create a ai message manually
ai_msg = AIMessage("lets solve your query")

# add conversation in history
conversation = [
    SystemMessage(content="you are a helpful assistant"),
    HumanMessage(content="can you help me"),
    ai_msg,  # insert if its a ai model response
    HumanMessage(content="great what is 132*132")
]
response = model.invoke(conversation)
print(response.content)
```

```python
from langchain.messages import ToolMessage

# AIMessage and HumanMessage are already imported in other cells
ai_message = AIMessage(
    content="calling tool get_weather",
    tool_calls=[{
        "name": "get_weather",
        "args": {"location": "San Francisco"},
        "id": "call_123"
    }]
)
weather_result = "sunny, 72 F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"
)
message = [
    HumanMessage(content="what the weather in San Francisco"),
    ai_message,
    tool_message,
]
response = model.invoke(message)
print(response)
```
